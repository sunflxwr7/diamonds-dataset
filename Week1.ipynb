```python
import pandas as pd
import matplotlib.pyplot as plt

# B Load the dataset
url = "https://raw.githubusercontent.com/sunflxwr7/diamonds-dataset/refs/heads/main/diamonds.csv"
diamonds = pd.read_csv(url, index_col=0)

# C Display the first seven rows
print("C Display the first seven rows")
print("First 7 Rows:\n")
print(diamonds.head(7))
print("\n" + "-"*50 + "\n")

# D Display the last seven rows
print("D Display the last seven rows")
print("Last 7 Rows:\n")
print(diamonds.tail(7))
print("\n" + "-"*50 + "\n")

# E Descriptive statistics for numerical columns
print("E Descriptive statistics for numerical columns")
print("Descriptive Statistics for Numerical Columns:\n")
print(diamonds.describe())
print("\n" + "-"*50 + "\n")

# F Descriptive statistics for categorical columns
print("F Descriptive statistics for categorical columns")
print("Descriptive Statistics for 'cut':\n")
print(diamonds['cut'].describe())
print("\nDescriptive Statistics for 'color':\n")
print(diamonds['color'].describe())
print("\nDescriptive Statistics for 'clarity':\n")
print(diamonds['clarity'].describe())
print("\n" + "-"*50 + "\n")

# G Unique values for categorical columns
print("G Unique values for categorical columns")
print("Unique Values in 'cut':", diamonds['cut'].unique())
print("Unique Values in 'color':", diamonds['color'].unique())
print("Unique Values in 'clarity':", diamonds['clarity'].unique())
print("\n" + "-"*50 + "\n")

#H Enable Matplotlib
%matplotlib inline

#H Plot histograms for all numerical columns
print("H Plot histograms for all numerical columns")
diamonds.hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```
